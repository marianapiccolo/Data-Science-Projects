### Import Libraries and Databases

```python
import pandas as pd
import pathlib # Lets you go through all files in a folder
```

### Consolidate database

```python
months = {'jan' : 1, 'fev' : 2, 'mar' : 3, 'abr' : 4, 'mai' : 5, 'jun' : 6, 'jul' : 7, 'ago' : 8, 'set' : 9, 'out' : 10, 'nov' : 11, 'dez' : 12}

base_path = pathlib.Path('dataset')

base_airbnb = pd.DataFrame()

for file in base_path.iterdir(): #iterdir = list with files in the folder
    name_month = file.name[:3] # 3 firts letters of the month
    month = months[name_month] # number of the month
    
    year = file.name[-8:] #20XX.csv
    year = int(year.replace('.csv', ''))
    
    df = pd.read_csv(base_path / file.name) # / = concatenate. Or: r'dataset\{}'.format(file.name)
    df['year'] = year #add column year
    df['month'] = month #add column month
    base_airbnb = base_airbnb.append(df)
    
display(base_airbnb)
```

### After qualitative analysis of the columns, taking into account the criteria explained above, we are left with the following columns:

```python
# Names pasted from Excel. See images 1, 2 and 3.
columns = ['host_response_time','host_response_rate','host_is_superhost','host_listings_count','latitude','longitude','property_type','room_type','accommodates','bathrooms','bedrooms','beds','bed_type','amenities','price','security_deposit','cleaning_fee','guests_included','extra_people','minimum_nights','maximum_nights','number_of_reviews','review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value','instant_bookable','is_business_travel_ready','cancellation_policy','year','month']

base_airbnb = base_airbnb.loc[:, columns]
display(base_airbnb)
```

### Handle missing values

```python
#print(base_airbnb.isnull().sum())

# Delete all columns with more than 300000 nulls.
for column in base_airbnb:
    if base_airbnb[column].isnull().sum() > 300000:
        base_airbnb = base_airbnb.drop(column, axis = 1)
        
#print(base_airbnb.isnull().sum())
#print(base_airbnb.shape)

# For the other columns, as there is a lot of data (more than 900000 rows), I will delete the rows with NaN.
base_airbnb = base_airbnb.dropna()
print(base_airbnb.isnull().sum())
print(base_airbnb.shape)
```
